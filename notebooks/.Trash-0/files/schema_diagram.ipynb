```python
import pandas as pd
from graphviz import Digraph

# Create a sample dataframe to represent the star schema
data = {
    'fact_table': ['Sales', 'Sales', 'Sales', 'Sales'],  # Add four 'Sales' entries
    'dimension_tables': ['Time', 'Product', 'Customer', 'Store']
}
df = pd.DataFrame(data)

# Create a graph object
dot = Digraph(comment='Star Schema')

# Add the fact table node
dot.node('fact_table', label='Sales')

# Add the dimension table nodes
for dim_table in df['dimension_tables']:
    dot.node(dim_table, label=dim_table)

# Add edges between the fact table and dimension tables
for dim_table in df['dimension_tables']:
    dot.edge('fact_table', dim_table)

# Render the graph
dot.render('star_schema', view=True)
```
