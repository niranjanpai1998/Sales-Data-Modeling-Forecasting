```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
df = pd.read_csv("data/sales_data_sample.csv", encoding="ISO-8859-1")
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df = df.sort_values(by='ORDERDATE')
```

```python
features = ['QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'MONTH_ID', 'QTR_ID', 'YEAR_ID', 'MSRP']
target = 'SALES'
```

```python
data = df[features + [target]]
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
train_size = int(len(data_scaled) * 0.8)
train_data = data_scaled[:train_size]
test_data = data_scaled[train_size:]
```

```python
seq_length=10
```

```python
def create_sequences(data, seq_length=10):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :-1])  # All features except SALES
        y.append(data[i+seq_length, -1])  # SALES as target
    return np.array(X), np.array(y)
```

```python
X_train, y_train = create_sequences(train_data, seq_length)
X_test, y_test = create_sequences(test_data, seq_length)
```

```python
model = Sequential([
    LSTM(50, activation='relu', return_sequences=True, input_shape=(seq_length, len(features))),
    LSTM(50, activation='relu'),
    Dense(25, activation='relu'),
    Dense(1)
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test_actual = scaler.inverse_transform(np.c_[np.zeros((y_test.shape[0], len(features))), y_test])[:, -1]
y_pred_actual = scaler.inverse_transform(np.c_[np.zeros((y_pred.shape[0], len(features))), y_pred])[:, -1]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y_test_actual, label='Actual Sales')
plt.plot(y_pred_actual, label='Predicted Sales', linestyle='dashed')
plt.legend()
plt.title('Actual vs Predicted Sales')
plt.xlabel('Time Steps')
plt.ylabel('Sales')
plt.show()
```
